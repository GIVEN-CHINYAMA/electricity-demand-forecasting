## 2. Problem Statement

Electricity grid operators face a critical operational challenge:

> **How much electricity will be consumed over the next 24 hours, 7 days,
> and 30 days — and when will demand peak?**

Getting this wrong in either direction is costly:

| Scenario | Consequence |
|----------|-------------|
| Demand HIGHER than forecast | Blackouts, grid instability, emergency purchases at high cost |
| Demand LOWER than forecast | Wasted generation, financial losses, environmental cost |

Traditional approaches rely on simple historical averages and expert
judgement. This project tests whether modern statistical and machine
learning models can produce significantly more accurate forecasts,
thereby reducing operational risk and cost.

### Business Questions This Project Answers

1. What are the dominant demand patterns (daily, weekly, annual)?
2. Is electricity demand stationary or does it trend over time?
3. Which model produces the most accurate short-term forecast?
4. How far ahead can we forecast with acceptable accuracy?
5. What practical recommendations follow from the forecast results?

## 3. Research Objectives

### Primary Objective
To build, evaluate, and compare ARIMA, SARIMA, Prophet, and LSTM models
for forecasting electricity demand, and to identify the most accurate
model for short-term operational forecasting.

### Specific Objectives

1. To collect and preprocess a real-world hourly/daily electricity
   demand dataset with sufficient length for time series analysis.

2. To perform comprehensive Exploratory Data Analysis including trend
   detection, seasonality analysis, and anomaly identification.

3. To engineer relevant time features (hour, day, month, lag variables,
   rolling averages) that improve model performance.

4. To test stationarity using the Augmented Dickey-Fuller (ADF) test
   and apply appropriate transformations.

5. To build and optimise ARIMA and SARIMA models using ACF/PACF
   analysis and automated parameter selection.

6. To build and tune a Prophet model incorporating seasonality
   components and holiday effects.

7. To build and train an LSTM neural network for sequence-based
   electricity demand forecasting.

8. To evaluate all four models using MAE, RMSE, MAPE, and R2 score
   on a held-out test set.

9. To generate a 30-day ahead forecast using the best-performing model.

10. To provide actionable recommendations for energy planners and
    grid operators based on the forecast results.

```python
# ============================================================
# CODE CELL 2.3 - Load Data and Run Validation Report
# ============================================================

# Load the CSV
df_raw = pd.read_csv('AEP_hourly.csv')

print("Raw data loaded.")
print(f"Shape         : {df_raw.shape}")
print(f"Columns       : {list(df_raw.columns)}")
print()
print(df_raw.head(5))
print()

# Parse datetime and set as index
df_raw.columns = ['Datetime', 'Demand_MW']
df_raw['Datetime'] = pd.to_datetime(df_raw['Datetime'])
df_raw = df_raw.sort_values('Datetime').drop_duplicates('Datetime')
df_raw = df_raw.set_index('Datetime')

# Full validation report
print("=" * 55)
print("  DATA VALIDATION REPORT")
print("=" * 55)
print(f"  Total hourly records : {len(df_raw):,}")
print(f"  Start date           : {df_raw.index.min()}")
print(f"  End date             : {df_raw.index.max()}")
print(f"  Missing values       : {df_raw['Demand_MW'].isna().sum()}")
print(f"  Min demand           : {df_raw['Demand_MW'].min():,.1f} MW")
print(f"  Max demand           : {df_raw['Demand_MW'].max():,.1f} MW")
print(f"  Mean demand          : {df_raw['Demand_MW'].mean():,.1f} MW")
print(f"  Std deviation        : {df_raw['Demand_MW'].std():,.1f} MW")
print("=" * 55)
print("Data loaded successfully.")
```

```python
# ============================================================
# CODE CELL 2.4 - Resample Hourly to Daily Averages
#
# Why resample?
# - Daily data is cleaner for ARIMA and SARIMA
# - Reduces noise while preserving trend and seasonality
# - 3,000+ daily rows are sufficient for all four models
# ============================================================

# Resample: take daily mean of hourly demand
df = df_raw['Demand_MW'].resample('D').mean()
df = df.dropna()
df = df.to_frame()

# Forward-fill any remaining gaps (public holidays etc.)
df['Demand_MW'] = df['Demand_MW'].ffill()

print("=" * 55)
print("  DAILY RESAMPLED DATA REPORT")
print("=" * 55)
print(f"  Total daily records  : {len(df):,}")
print(f"  Start date           : {df.index.min().date()}")
print(f"  End date             : {df.index.max().date()}")
print(f"  Missing values       : {df['Demand_MW'].isna().sum()}")
print(f"  Min daily demand     : {df['Demand_MW'].min():,.1f} MW")
print(f"  Max daily demand     : {df['Demand_MW'].max():,.1f} MW")
print(f"  Mean daily demand    : {df['Demand_MW'].mean():,.1f} MW")
print("=" * 55)

# Save checkpoint
df.to_csv('outputs/electricity_demand_daily.csv')
print("Daily data saved to: outputs/electricity_demand_daily.csv")
print()
df.tail(5)
```

## Stage 4 - Feature Engineering

Feature engineering transforms raw data into inputs that help models
learn patterns more effectively. We create three types of features:

1. Calendar features - hour, day, month, season, weekend flag
2. Lag features - yesterday's demand, last week, last month
3. Rolling window features - 7-day mean, 30-day mean, 7-day std

These features are primarily used by the LSTM and can also provide
context for evaluating ARIMA/SARIMA residuals.

```python
# ============================================================
# CODE CELL 4.1 - Feature Engineering
# ============================================================

# Work on a copy to preserve the original df
df_feat = df[['Demand_MW']].copy()

# --- 1. Calendar features ---
df_feat['Year']       = df_feat.index.year
df_feat['Month']      = df_feat.index.month
df_feat['DayOfWeek']  = df_feat.index.dayofweek   # 0=Mon, 6=Sun
df_feat['Quarter']    = df_feat.index.quarter
df_feat['IsWeekend']  = (df_feat['DayOfWeek'] >= 5).astype(int)
df_feat['DayOfYear']  = df_feat.index.dayofyear

# Season: 1=Winter, 2=Spring, 3=Summer, 4=Autumn
def get_season(month):
    if month in [12, 1, 2]:  return 1   # Winter
    elif month in [3, 4, 5]: return 2   # Spring
    elif month in [6, 7, 8]: return 3   # Summer
    else:                    return 4   # Autumn

df_feat['Season'] = df_feat['Month'].apply(get_season)
season_map = {1:'Winter', 2:'Spring', 3:'Summer', 4:'Autumn'}

# --- 2. Lag features ---
df_feat['Lag_1']   = df_feat['Demand_MW'].shift(1)    # Yesterday
df_feat['Lag_7']   = df_feat['Demand_MW'].shift(7)    # Same day last week
df_feat['Lag_30']  = df_feat['Demand_MW'].shift(30)   # Same day last month
df_feat['Lag_365'] = df_feat['Demand_MW'].shift(365)  # Same day last year

# --- 3. Rolling window features ---
df_feat['Rolling_7_mean']  = df_feat['Demand_MW'].rolling(7).mean()
df_feat['Rolling_30_mean'] = df_feat['Demand_MW'].rolling(30).mean()
df_feat['Rolling_7_std']   = df_feat['Demand_MW'].rolling(7).std()
df_feat['Rolling_30_std']  = df_feat['Demand_MW'].rolling(30).std()

# Drop rows with NaN from lag/rolling (first 365 rows)
df_feat_clean = df_feat.dropna()

print("=" * 55)
print("  FEATURE ENGINEERING REPORT")
print("=" * 55)
print(f"  Original rows        : {len(df):,}")
print(f"  After lag drop (365) : {len(df_feat_clean):,}")
print(f"  Total features       : {df_feat_clean.shape[1]}")
print(f"  Feature columns      :")
for col in df_feat_clean.columns:
    print(f"    - {col}")
print("=" * 55)

# Save
df_feat_clean.to_csv('outputs/electricity_features_engineered.csv')
print("\nSaved: outputs/electricity_features_engineered.csv")
```

## Stage 5 - Stationarity Testing and Time Series Decomposition

Before fitting ARIMA or SARIMA we must verify stationarity and
understand the decomposed components of the series:

1. Seasonal decomposition (Trend + Seasonality + Residuals)
2. Augmented Dickey-Fuller (ADF) stationarity test
3. Differencing to achieve stationarity
4. ACF and PACF plots to identify ARIMA order (p, d, q)

```python
# ============================================================
# CODE CELL 5.1 - Seasonal Decomposition
# Breaks the series into: Trend + Seasonality + Residuals
# ============================================================

# Use monthly resampled data for cleaner decomposition
monthly = df['Demand_MW'].resample('M').mean()

decomp = seasonal_decompose(monthly, model='additive', period=12)

fig, axes = plt.subplots(4, 1, figsize=(16, 14))

decomp.observed.plot(ax=axes[0], color='steelblue',   linewidth=1.3)
axes[0].set_title('Observed (Original Series)',         fontweight='bold')
axes[0].set_ylabel('MW')

decomp.trend.plot(ax=axes[1],    color='red',           linewidth=1.5)
axes[1].set_title('Trend Component',                    fontweight='bold')
axes[1].set_ylabel('MW')

decomp.seasonal.plot(ax=axes[2], color='green',         linewidth=1.2)
axes[2].set_title('Seasonal Component (Annual Pattern)', fontweight='bold')
axes[2].set_ylabel('MW')

decomp.resid.plot(ax=axes[3],    color='gray',          linewidth=0.8)
axes[3].axhline(0, color='black', linestyle='--', linewidth=1)
axes[3].set_title('Residual Component (Noise)',          fontweight='bold')
axes[3].set_ylabel('MW')
axes[3].set_xlabel('Date')

for ax in axes:
    ax.yaxis.set_major_formatter(
        mticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))

fig.suptitle('Seasonal Decomposition of Monthly Electricity Demand',
             fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('outputs/11_decomposition.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: outputs/11_decomposition.png")
print()
print("INTERPRETATION:")
print("  - TREND: Clear peak ~2007, decline through 2009, recovery 2013+")
print("  - SEASONAL: Consistent 12-month cycle with summer/winter peaks")
print("  - RESIDUALS: Mostly random noise — good sign for ARIMA modelling")
```

```python
# ============================================================
# CODE CELL 5.2 - Augmented Dickey-Fuller Stationarity Tests
# ============================================================

def adf_test(series, name):
    result = adfuller(series.dropna(), autolag='AIC')
    print(f"\n{'='*55}")
    print(f"  ADF Test: {name}")
    print(f"{'='*55}")
    print(f"  ADF Statistic  : {result[0]:.6f}")
    print(f"  p-value        : {result[1]:.6f}")
    print(f"  Lags Used      : {result[2]}")
    for level, cval in result[4].items():
        print(f"  Critical {level}  : {cval:.4f}")
    print()
    if result[1] <= 0.05:
        print(f"  VERDICT: STATIONARY (p <= 0.05) ✅")
    else:
        print(f"  VERDICT: NON-STATIONARY (p > 0.05) ❌")
    return result[1]

# Test 1: Original daily series
p1 = adf_test(df['Demand_MW'], "Original Daily Series")

# Test 2: First difference
df['Demand_diff1'] = df['Demand_MW'].diff()
p2 = adf_test(df['Demand_diff1'], "First Difference (d=1)")

# Test 3: Seasonal difference (lag-365)
df['Demand_sdiff'] = df['Demand_MW'].diff(365)
p3 = adf_test(df['Demand_sdiff'], "Seasonal Difference (lag-365)")

# Test 4: Both differences
df['Demand_both'] = df['Demand_MW'].diff().diff(365)
p4 = adf_test(df['Demand_both'], "First + Seasonal Difference")

print()
print("=" * 55)
print("  SUMMARY")
print("=" * 55)
print(f"  Original series      : p={p1:.4f} {'✅' if p1<=0.05 else '❌'}")
print(f"  First difference     : p={p2:.4f} {'✅' if p2<=0.05 else '❌'}")
print(f"  Seasonal difference  : p={p3:.4f} {'✅' if p3<=0.05 else '❌'}")
print(f"  Both differences     : p={p4:.4f} {'✅' if p4<=0.05 else '❌'}")
```

```python
# ============================================================
# CODE CELL 5.3 - ACF and PACF Plots
# ACF  -> identifies MA order q
# PACF -> identifies AR order p
# ============================================================

# Use first-differenced series (confirmed stationary)
diff_series = df['Demand_MW'].diff().dropna()

fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# ACF - original
plot_acf(df['Demand_MW'].dropna(), lags=60, ax=axes[0,0],
         title='ACF - Original Series', color='steelblue')

# PACF - original
plot_pacf(df['Demand_MW'].dropna(), lags=60, ax=axes[0,1],
          title='PACF - Original Series', color='steelblue', method='ywm')

# ACF - differenced
plot_acf(diff_series, lags=60, ax=axes[1,0],
         title='ACF - First Differenced Series (Stationary)',
         color='darkorange')

# PACF - differenced
plot_pacf(diff_series, lags=60, ax=axes[1,1],
          title='PACF - First Differenced Series (Stationary)',
          color='darkorange', method='ywm')

for ax in axes.flat:
    ax.axhline(y=0, color='black', linewidth=0.8)

plt.suptitle('ACF and PACF Analysis — Electricity Demand',
             fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('outputs/12_acf_pacf.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: outputs/12_acf_pacf.png")
print()
print("HOW TO READ THESE PLOTS:")
print("  ACF cuts off at lag k  -> MA order  q = k")
print("  PACF cuts off at lag k -> AR order  p = k")
print("  Spikes at lag 7/365    -> seasonal pattern confirmed")
print()
print("  auto_arima in Stage 6 will confirm exact (p,d,q) values.")

# Save d value for ARIMA
d_value = 1
print(f"\n  Confirmed: d = {d_value} (first differencing achieves stationarity)")
```

## Stage 6 - ARIMA Model

ARIMA (AutoRegressive Integrated Moving Average) is the classical
statistical baseline for time series forecasting.

The ADF test confirmed our series is already stationary (p=0.0000),
meaning d=0 is valid. We use auto_arima to find the optimal (p,d,q).

Steps:
1. auto_arima selects best (p,d,q) using AIC criterion
2. Fit ARIMA on training data
3. Diagnostic check of residuals
4. Forecast on test set
5. Evaluate with RMSE, MAE, MAPE

```python
# ============================================================
# CODE CELL 6.1 - Auto ARIMA: Find Best (p, d, q)
# This searches all combinations and picks lowest AIC
# May take 3-5 minutes — please wait
# ============================================================

print("Running auto_arima on training data...")
print("Searching p=0..3, d=0..1, q=0..3 ...")
print()

arima_search = auto_arima(
    train['Demand_MW'],
    start_p=0, max_p=3,
    start_q=0, max_q=3,
    d=None,
    seasonal=False,
    information_criterion='aic',
    trace=True,
    error_action='ignore',
    suppress_warnings=True,
    stepwise=True,
    random_state=42
)

p, d, q = arima_search.order

print()
print("=" * 55)
print("  AUTO ARIMA RESULT")
print("=" * 55)
print(f"  Best order  : ARIMA({p}, {d}, {q})")
print(f"  AIC score   : {arima_search.aic():.4f}")
print(f"  BIC score   : {arima_search.bic():.4f}")
print("=" * 55)
print()
print("  Lower AIC/BIC = better model fit")
print(f"  Proceeding to fit ARIMA({p},{d},{q})")
```

```python
# ============================================================
# CODE CELL 6.2 - Fit Final ARIMA Model
# ============================================================

print(f"Fitting ARIMA({p},{d},{q}) on {len(train):,} training rows...")

arima_model = ARIMA(train['Demand_MW'], order=(p, d, q)).fit()

print(arima_model.summary())
print()
print("=" * 55)
print("  ARIMA MODEL DIAGNOSTICS")
print("=" * 55)
print(f"  Order   : ARIMA({p},{d},{q})")
print(f"  AIC     : {arima_model.aic:.4f}")
print(f"  BIC     : {arima_model.bic:.4f}")
print(f"  HQIC    : {arima_model.hqic:.4f}")
print("=" * 55)
```

```python
# ============================================================
# CODE CELL 6.3 - ARIMA Residual Diagnostics
# Good residuals = white noise (random, no patterns)
# ============================================================

residuals = arima_model.resid

fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# 1. Residuals over time
axes[0,0].plot(residuals.index, residuals.values,
               color='steelblue', linewidth=0.8)
axes[0,0].axhline(0, color='red', linestyle='--', linewidth=1)
axes[0,0].set_title('Residuals Over Time', fontweight='bold')
axes[0,0].set_ylabel('Residual (MW)')
axes[0,0].yaxis.set_major_formatter(
    mticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))

# 2. Residuals histogram
axes[0,1].hist(residuals, bins=60, color='steelblue',
               edgecolor='white', alpha=0.85)
axes[0,1].axvline(0, color='red', linestyle='--', linewidth=1.5)
axes[0,1].set_title('Residuals Distribution', fontweight='bold')
axes[0,1].set_xlabel('Residual Value (MW)')
axes[0,1].set_ylabel('Frequency')

# 3. ACF of residuals (should all be near zero)
plot_acf(residuals, lags=40, ax=axes[1,0],
         title='ACF of Residuals\n(bars should be within blue bands)',
         color='steelblue')

# 4. Q-Q plot (points should follow the diagonal line)
from statsmodels.graphics.gofplots import qqplot
qqplot(residuals, line='s', ax=axes[1,1], alpha=0.4)
axes[1,1].set_title('Q-Q Plot of Residuals\n(should follow diagonal)',
                    fontweight='bold')

plt.suptitle(f'ARIMA({p},{d},{q}) Residual Diagnostics',
             fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('outputs/14_arima_diagnostics.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: outputs/14_arima_diagnostics.png")
print()
print("WHAT TO LOOK FOR:")
print("  Residuals over time  : Random scatter around zero = GOOD")
print("  Histogram            : Bell-shaped, centred at zero = GOOD")
print("  ACF of residuals     : All bars within blue bands = GOOD")
print("  Q-Q plot             : Points on diagonal line = GOOD")
```

```python
# ============================================================
# CODE CELL 6.4 - ARIMA Forecast on Test Set
# ============================================================

n_test = len(test)
print(f"Generating ARIMA({p},{d},{q}) forecast for {n_test} test days...")

forecast_obj    = arima_model.get_forecast(steps=n_test)
arima_pred      = forecast_obj.predicted_mean
forecast_ci     = forecast_obj.conf_int(alpha=0.05)
ci_lower        = forecast_ci.iloc[:, 0]
ci_upper        = forecast_ci.iloc[:, 1]

# Align index with test dates
arima_pred.index = test.index
ci_lower.index   = test.index
ci_upper.index   = test.index

# Plot
fig, ax = plt.subplots(figsize=(16, 6))
ax.plot(train['Demand_MW'].iloc[-120:],
        label='Training (last 120 days)',
        color='steelblue', linewidth=1.2)
ax.plot(test['Demand_MW'],
        label='Actual (Test Set)',
        color='black', linewidth=1.5)
ax.plot(test.index, arima_pred,
        label=f'ARIMA({p},{d},{q}) Forecast',
        color='red', linewidth=1.5, linestyle='--')
ax.fill_between(test.index, ci_lower, ci_upper,
                alpha=0.12, color='red',
                label='95% Confidence Interval')
ax.axvline(test.index[0], color='gray',
           linestyle=':', linewidth=1.2)
ax.set_title(f'ARIMA({p},{d},{q}) — Forecast vs Actual Electricity Demand',
             fontweight='bold', fontsize=13)
ax.set_ylabel('Demand (MW)')
ax.legend(loc='upper left')
ax.yaxis.set_major_formatter(
    mticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))

plt.tight_layout()
plt.savefig('outputs/15_arima_forecast.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: outputs/15_arima_forecast.png")
```

```python
# ============================================================
# CODE CELL 6.5 - ARIMA Evaluation Metrics
# ============================================================

actual      = test['Demand_MW'].values
arima_vals  = arima_pred.values

arima_rmse  = np.sqrt(mean_squared_error(actual, arima_vals))
arima_mae   = mean_absolute_error(actual, arima_vals)
arima_mape  = np.mean(np.abs((actual - arima_vals) / actual)) * 100
arima_r2    = r2_score(actual, arima_vals)

print("=" * 55)
print(f"  ARIMA({p},{d},{q}) — TEST SET PERFORMANCE")
print("=" * 55)
print(f"  RMSE  : {arima_rmse:,.4f} MW")
print(f"  MAE   : {arima_mae:,.4f}  MW")
print(f"  MAPE  : {arima_mape:.2f}%")
print(f"  R²    : {arima_r2:.4f}")
print("=" * 55)
print()
print("  RMSE = Root Mean Squared Error")
print("  MAE  = Mean Absolute Error")
print("  MAPE = Mean Absolute Percentage Error (% accuracy)")
print("  R²   = Proportion of variance explained (1.0 = perfect)")

# Save for Stage 10 comparison
arima_results = {
    'Model' : f'ARIMA({p},{d},{q})',
    'RMSE'  : round(arima_rmse, 2),
    'MAE'   : round(arima_mae,  2),
    'MAPE'  : round(arima_mape, 2),
    'R2'    : round(arima_r2,   4)
}
print()
print("ARIMA results saved for Stage 10 comparison.")
```

## Stage 7 - SARIMA Model

SARIMA extends ARIMA by adding seasonal components.
It is defined by two sets of parameters:

- Non-seasonal : (p, d, q)
- Seasonal     : (P, D, Q, m)  where m = seasonal period

Since our data shows clear annual seasonality and we are working
with daily data resampled to capture weekly cycles, we use m=7
(weekly seasonality) which is computationally feasible.

auto_arima with seasonal=True will find the best full order.

```python
# ============================================================
# CODE CELL 7.1 - Auto SARIMA Order Selection
# Uses weekly seasonality m=7 (computationally practical)
# This will take 5-10 minutes — please wait
# ============================================================

print("Running auto_arima with seasonal=True (m=7)...")
print("This searches seasonal (P,D,Q) on top of (p,d,q)")
print("May take 5-10 minutes on Colab CPU...")
print()

sarima_search = auto_arima(
    train['Demand_MW'],
    start_p=0, max_p=2,
    start_q=0, max_q=2,
    d=None,
    start_P=0, max_P=2,
    start_Q=0, max_Q=2,
    D=None,
    m=7,
    seasonal=True,
    information_criterion='aic',
    trace=True,
    error_action='ignore',
    suppress_warnings=True,
    stepwise=True,
    random_state=42
)

sp, sd, sq, sm = (sarima_search.order[0],
                  sarima_search.order[1],
                  sarima_search.order[2],
                  7)
sP, sD, sQ     = sarima_search.seasonal_order[:3]

print()
print("=" * 55)
print("  AUTO SARIMA RESULT")
print("=" * 55)
print(f"  Non-seasonal : ({sp},{sd},{sq})")
print(f"  Seasonal     : ({sP},{sD},{sQ},{sm})")
print(f"  AIC          : {sarima_search.aic():.4f}")
print(f"  BIC          : {sarima_search.bic():.4f}")
print("=" * 55)
```

```python
# ============================================================
# CODE CELL 7.2 - Fit SARIMA and Generate Test Forecast
# ============================================================

print(f"Fitting SARIMA({sp},{sd},{sq})({sP},{sD},{sQ},{sm})...")

sarima_model = SARIMAX(
    train['Demand_MW'],
    order=(sp, sd, sq),
    seasonal_order=(sP, sD, sQ, sm),
    enforce_stationarity=False,
    enforce_invertibility=False
).fit(disp=False)

print(sarima_model.summary())
print()
print("=" * 55)
print("  SARIMA MODEL INFO")
print("=" * 55)
print(f"  Full order : SARIMA({sp},{sd},{sq})({sP},{sD},{sQ},{sm})")
print(f"  AIC        : {sarima_model.aic:.4f}")
print(f"  BIC        : {sarima_model.bic:.4f}")
print("=" * 55)

# Forecast
sarima_forecast  = sarima_model.get_forecast(steps=n_test)
sarima_pred      = sarima_forecast.predicted_mean
sarima_ci        = sarima_forecast.conf_int(alpha=0.05)
sarima_pred.index      = test.index
sarima_ci.index        = test.index

# Evaluation
sarima_vals  = sarima_pred.values
sarima_rmse  = np.sqrt(mean_squared_error(actual, sarima_vals))
sarima_mae   = mean_absolute_error(actual, sarima_vals)
sarima_mape  = np.mean(np.abs((actual - sarima_vals) / actual)) * 100
sarima_r2    = r2_score(actual, sarima_vals)

print()
print("=" * 55)
print(f"  SARIMA — TEST SET PERFORMANCE")
print("=" * 55)
print(f"  RMSE  : {sarima_rmse:,.4f} MW")
print(f"  MAE   : {sarima_mae:,.4f}  MW")
print(f"  MAPE  : {sarima_mape:.2f}%")
print(f"  R²    : {sarima_r2:.4f}")
print("=" * 55)

sarima_results = {
    'Model' : f'SARIMA({sp},{sd},{sq})({sP},{sD},{sQ},{sm})',
    'RMSE'  : round(sarima_rmse, 2),
    'MAE'   : round(sarima_mae,  2),
    'MAPE'  : round(sarima_mape, 2),
    'R2'    : round(sarima_r2,   4)
}
print("SARIMA results saved for Stage 10 comparison.")
```

```python
# ============================================================
# CODE CELL 7.3 - SARIMA Forecast Plot
# ============================================================

fig, axes = plt.subplots(2, 1, figsize=(16, 12))

# Top: ARIMA
axes[0].plot(train['Demand_MW'].iloc[-120:],
             label='Training (last 120 days)',
             color='steelblue', linewidth=1.2)
axes[0].plot(test['Demand_MW'],
             label='Actual', color='black', linewidth=1.5)
axes[0].plot(test.index, arima_pred,
             label=f'ARIMA({p},{d},{q})',
             color='red', linewidth=1.5, linestyle='--')
axes[0].fill_between(test.index,
                     ci_lower, ci_upper,
                     alpha=0.1, color='red', label='95% CI')
axes[0].set_title(f'ARIMA({p},{d},{q})  |  RMSE: {arima_rmse:,.2f} MW'
                  f'  |  MAPE: {arima_mape:.2f}%',
                  fontweight='bold', fontsize=12)
axes[0].set_ylabel('Demand (MW)')
axes[0].legend(loc='upper left', fontsize=9)
axes[0].yaxis.set_major_formatter(
    mticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))

# Bottom: SARIMA
axes[1].plot(train['Demand_MW'].iloc[-120:],
             label='Training (last 120 days)',
             color='steelblue', linewidth=1.2)
axes[1].plot(test['Demand_MW'],
             label='Actual', color='black', linewidth=1.5)
axes[1].plot(test.index, sarima_pred,
             label=f'SARIMA({sp},{sd},{sq})({sP},{sD},{sQ},{sm})',
             color='green', linewidth=1.5, linestyle='--')
axes[1].fill_between(test.index,
                     sarima_ci.iloc[:,0],
                     sarima_ci.iloc[:,1],
                     alpha=0.1, color='green', label='95% CI')
axes[1].set_title(f'SARIMA({sp},{sd},{sq})({sP},{sD},{sQ},{sm})'
                  f'  |  RMSE: {sarima_rmse:,.2f} MW'
                  f'  |  MAPE: {sarima_mape:.2f}%',
                  fontweight='bold', fontsize=12)
axes[1].set_ylabel('Demand (MW)')
axes[1].set_xlabel('Date')
axes[1].legend(loc='upper left', fontsize=9)
axes[1].yaxis.set_major_formatter(
    mticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))

plt.suptitle('ARIMA vs SARIMA — Electricity Demand Forecast',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('outputs/16_arima_sarima_comparison.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Saved: outputs/16_arima_sarima_comparison.png")
print()
print("FOR REFERENCE:")
print(f"  ARIMA  MAPE : {arima_mape:.2f}%")
print(f"  SARIMA MAPE : {sarima_mape:.2f}%")
winner = "SARIMA" if sarima_mape < arima_mape else "ARIMA"
print(f"  Winner so far: {winner}")
```

## Stage 8 - Prophet Model

Prophet is an open-source forecasting tool developed by Meta (Facebook).
It is designed specifically for business time series with:
- Strong seasonal patterns (daily, weekly, annual)
- Missing data and outliers
- Holiday effects
- Non-linear trends

Prophet requires the dataframe to have exactly two columns:
- ds : the date column
- y  : the value column

It automatically detects and models trend + seasonality + holidays.

```python
# ============================================================
# CODE CELL 8.1 - Prepare Data and Fit Prophet Model
# ============================================================

# Prophet requires columns named 'ds' and 'y' exactly
prophet_train = train[['Demand_MW']].reset_index()
prophet_train.columns = ['ds', 'y']

prophet_test  = test[['Demand_MW']].reset_index()
prophet_test.columns  = ['ds', 'y']

print(f"Training rows : {len(prophet_train):,}")
print(f"Test rows     : {len(prophet_test):,}")
print()
print("Fitting Prophet model...")

prophet_model = Prophet(
    yearly_seasonality  = True,
    weekly_seasonality  = True,
    daily_seasonality   = False,
    seasonality_mode    = 'additive',
    changepoint_prior_scale = 0.05,
    seasonality_prior_scale = 10.0,
    interval_width      = 0.95
)

prophet_model.fit(prophet_train)

print()
print("=" * 55)
print("  PROPHET MODEL CONFIGURATION")
print("=" * 55)
print("  Yearly seasonality  : True")
print("  Weekly seasonality  : True")
print("  Daily seasonality   : False")
print("  Seasonality mode    : Additive")
print("  Changepoint scale   : 0.05")
print("=" * 55)
print("Prophet model fitted successfully.")
```

## Stage 9 - LSTM Model

LSTM (Long Short-Term Memory) is a deep learning model built for
sequential data. Unlike ARIMA/SARIMA/Prophet, LSTM:

- Requires no stationarity assumptions
- Can learn complex non-linear patterns
- Captures long-range dependencies via memory gates
- Uses MinMax scaling and sliding window sequences

Architecture:
- Input    : 60-day lookback window
- Layer 1  : LSTM(64 units) + Dropout(0.2)
- Layer 2  : LSTM(32 units) + Dropout(0.2)
- Layer 3  : Dense(16, ReLU)
- Output   : Dense(1, Linear) — next day demand
- Optimizer: Adam  |  Loss: MSE

```python
# ============================================================
# CODE CELL 9.1 - Scale Data and Create LSTM Sequences
# IMPORTANT: Fit scaler on train only to prevent data leakage
# ============================================================

LOOKBACK = 60  # Model looks back 60 days to predict the next day

# Scale: fit on train, transform both
scaler       = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train[['Demand_MW']])
test_scaled  = scaler.transform(test[['Demand_MW']])

def create_sequences(data, lookback):
    """Create sliding window (X, y) pairs for LSTM"""
    X, y = [], []
    for i in range(lookback, len(data)):
        X.append(data[i-lookback:i, 0])  # past 60 days
        y.append(data[i, 0])              # next day
    return np.array(X), np.array(y)

X_train, y_train = create_sequences(train_scaled, LOOKBACK)

# Prepend last LOOKBACK rows of train so test has full context
combined         = np.concatenate([train_scaled[-LOOKBACK:], test_scaled])
X_test,  y_test  = create_sequences(combined, LOOKBACK)

# Reshape to [samples, timesteps, features]
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test  = X_test.reshape ((X_test.shape[0],  X_test.shape[1],  1))

print("=" * 55)
print("  LSTM DATA PREPARATION")
print("=" * 55)
print(f"  Lookback window  : {LOOKBACK} days")
print(f"  Scaler range     : [0, 1]")
print(f"  X_train shape    : {X_train.shape}")
print(f"  y_train shape    : {y_train.shape}")
print(f"  X_test  shape    : {X_test.shape}")
print(f"  y_test  shape    : {y_test.shape}")
print("=" * 55)
print("Data ready for LSTM training.")
```

```python
# ============================================================
# CODE CELL 9.2 - Build LSTM Neural Network
# ============================================================

tf.random.set_seed(SEED)
np.random.seed(SEED)

lstm_model = Sequential([
    LSTM(64, return_sequences=True,
         input_shape=(LOOKBACK, 1),
         name='lstm_layer_1'),
    Dropout(0.2, name='dropout_1'),

    LSTM(32, return_sequences=False,
         name='lstm_layer_2'),
    Dropout(0.2, name='dropout_2'),

    Dense(16, activation='relu', name='dense_hidden'),
    Dense(1,  activation='linear', name='output')
])

lstm_model.compile(
    optimizer='adam',
    loss='mean_squared_error',
    metrics=['mae']
)

print("=" * 55)
print("  LSTM ARCHITECTURE")
print("=" * 55)
lstm_model.summary()
print()
print(f"  Total parameters : {lstm_model.count_params():,}")
print("  Model built successfully.")
```

```python
# ============================================================
# CODE CELL 9.3 - Train the LSTM Model
# Early stopping prevents overfitting
# Expected time: 5-10 minutes on Colab CPU
# ============================================================

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=7,
    min_lr=1e-6,
    verbose=1
)

print("Starting LSTM training...")
print(f"Max epochs: 150  |  Batch: 32  |  Validation: 10%")
print()

history = lstm_model.fit(
    X_train, y_train,
    epochs=150,
    batch_size=32,
    validation_split=0.10,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)

best_epoch = np.argmin(history.history['val_loss']) + 1
print()
print("=" * 55)
print("  TRAINING COMPLETE")
print("=" * 55)
print(f"  Best epoch       : {best_epoch}")
print(f"  Final train loss : {history.history['loss'][-1]:.6f}")
print(f"  Final val loss   : {history.history['val_loss'][-1]:.6f}")
print("=" * 55)
```

```python
# ============================================================
# CODE CELL 9.4 - Plot LSTM Training Loss Curves
# ============================================================

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Loss
axes[0].plot(history.history['loss'],
             label='Training Loss', color='steelblue', linewidth=1.5)
axes[0].plot(history.history['val_loss'],
             label='Validation Loss', color='darkorange', linewidth=1.5)
axes[0].axvline(best_epoch-1, color='red', linestyle='--',
                alpha=0.7, label=f'Best Epoch ({best_epoch})')
axes[0].set_title('LSTM Training vs Validation Loss (MSE)',
                  fontweight='bold', fontsize=13)
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('MSE Loss')
axes[0].legend()

# MAE
axes[1].plot(history.history['mae'],
             label='Training MAE', color='steelblue', linewidth=1.5)
axes[1].plot(history.history['val_mae'],
             label='Validation MAE', color='darkorange', linewidth=1.5)
axes[1].set_title('LSTM Training vs Validation MAE',
                  fontweight='bold', fontsize=13)
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('MAE Loss (scaled)')
axes[1].legend()

plt.tight_layout()
plt.savefig('outputs/19_lstm_training_loss.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: outputs/19_lstm_training_loss.png")
```

```python
# ============================================================
# CODE CELL 9.5 - Generate Predictions and Evaluate LSTM
# ============================================================

# Predict on test set (output is scaled [0,1])
lstm_pred_scaled = lstm_model.predict(X_test, verbose=0)

# Inverse transform back to MW
lstm_pred  = scaler.inverse_transform(lstm_pred_scaled).flatten()
actual_inv = scaler.inverse_transform(y_test.reshape(-1,1)).flatten()

# Evaluation
lstm_rmse  = np.sqrt(mean_squared_error(actual_inv, lstm_pred))
lstm_mae   = mean_absolute_error(actual_inv, lstm_pred)
lstm_mape  = np.mean(np.abs((actual_inv - lstm_pred) / actual_inv)) * 100
lstm_r2    = r2_score(actual_inv, lstm_pred)

print("=" * 55)
print("  LSTM — TEST SET PERFORMANCE")
print("=" * 55)
print(f"  RMSE  : {lstm_rmse:,.4f} MW")
print(f"  MAE   : {lstm_mae:,.4f}  MW")
print(f"  MAPE  : {lstm_mape:.2f}%")
print(f"  R²    : {lstm_r2:.4f}")
print("=" * 55)
print()
print("  RUNNING SCOREBOARD:")
print(f"  ARIMA   MAPE : {arima_mape:.2f}%")
print(f"  SARIMA  MAPE : {sarima_mape:.2f}%")
print(f"  Prophet MAPE : {prophet_mape:.2f}%")
print(f"  LSTM    MAPE : {lstm_mape:.2f}%")

lstm_results = {
    'Model' : 'LSTM',
    'RMSE'  : round(lstm_rmse, 2),
    'MAE'   : round(lstm_mae,  2),
    'MAPE'  : round(lstm_mape, 2),
    'R2'    : round(lstm_r2,   4)
}
print()
print("LSTM results saved for Stage 10 comparison.")
```

## Stage 11 - 30-Day Future Forecast

Using the best performing model (LSTM) and all other models
re-fitted on the FULL dataset (2004-2018), we generate a
30-business-day forecast beyond August 2018.

- ARIMA/SARIMA: one-shot forecast with confidence intervals
- Prophet: extends its learned trend and seasonality
- LSTM: rolling one-step-ahead forecast

```python
# ============================================================
# CODE CELL 11.1 - ARIMA & SARIMA 30-Day Future Forecast
# Re-fit on FULL dataset for best future forecast
# ============================================================

N_DAYS = 30
future_dates = pd.bdate_range(
    start=df.index[-1] + pd.Timedelta(days=1),
    periods=N_DAYS
)

print(f"Forecasting {N_DAYS} days from {future_dates[0].date()} to {future_dates[-1].date()}")
print()

# --- ARIMA on full data ---
print(f"Re-fitting ARIMA({p},{d},{q}) on full dataset...")
arima_final     = ARIMA(df['Demand_MW'], order=(p,d,q)).fit()
arima_fut_obj   = arima_final.get_forecast(steps=N_DAYS)
arima_fut       = arima_fut_obj.predicted_mean
arima_fut_ci    = arima_fut_obj.conf_int(alpha=0.05)
arima_fut.index = future_dates
print("ARIMA future forecast done.")

# --- SARIMA on full data ---
print(f"Re-fitting SARIMA({sp},{sd},{sq})({sP},{sD},{sQ},{sm}) on full dataset...")
sarima_final    = SARIMAX(df['Demand_MW'],
                          order=(sp,sd,sq),
                          seasonal_order=(sP,sD,sQ,sm),
                          enforce_stationarity=False,
                          enforce_invertibility=False).fit(disp=False)
sarima_fut_obj  = sarima_final.get_forecast(steps=N_DAYS)
sarima_fut      = sarima_fut_obj.predicted_mean
sarima_fut.index= future_dates
print("SARIMA future forecast done.")

print(f"\nARIMA  30-day avg forecast : {arima_fut.mean():,.1f} MW")
print(f"SARIMA 30-day avg forecast : {sarima_fut.mean():,.1f} MW")
```

```python
# ============================================================
# CODE CELL 11.2 - Prophet 30-Day Future Forecast
# ============================================================

print("Re-fitting Prophet on full dataset...")

prophet_full = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=False,
    seasonality_mode='additive',
    changepoint_prior_scale=0.05,
    interval_width=0.95
)

prophet_all = df[['Demand_MW']].reset_index()
prophet_all.columns = ['ds', 'y']
prophet_full.fit(prophet_all)

future_prophet = prophet_full.make_future_dataframe(
    periods=N_DAYS, freq='D'
)
forecast_prophet = prophet_full.predict(future_prophet)

prophet_fut       = forecast_prophet.tail(N_DAYS)['yhat'].values
prophet_fut_lower = forecast_prophet.tail(N_DAYS)['yhat_lower'].values
prophet_fut_upper = forecast_prophet.tail(N_DAYS)['yhat_upper'].values

print(f"Prophet 30-day avg forecast : {prophet_fut.mean():,.1f} MW")
print("Prophet forecast done.")
```

```python
# ============================================================
# CODE CELL 11.3 - LSTM 30-Day Rolling Forecast
# Each prediction feeds back as input for the next step
# ============================================================

print("Re-fitting scaler and LSTM on full dataset...")

full_scaler  = MinMaxScaler(feature_range=(0,1))
full_scaled  = full_scaler.fit_transform(df[['Demand_MW']])

X_full, y_full = create_sequences(full_scaled, LOOKBACK)
X_full = X_full.reshape((X_full.shape[0], X_full.shape[1], 1))

lstm_model.fit(X_full, y_full,
               epochs=30, batch_size=32, verbose=0)
print("Re-training on full data done.")

# Rolling forecast
last_seq       = full_scaled[-LOOKBACK:].reshape(1, LOOKBACK, 1)
lstm_fut_scaled= []

for step in range(N_DAYS):
    next_pred = lstm_model.predict(last_seq, verbose=0)[0, 0]
    lstm_fut_scaled.append(next_pred)
    last_seq  = np.append(
        last_seq[0, 1:, :], [[next_pred]], axis=0
    ).reshape(1, LOOKBACK, 1)

lstm_fut = full_scaler.inverse_transform(
    np.array(lstm_fut_scaled).reshape(-1,1)
).flatten()

print(f"LSTM   30-day avg forecast : {lstm_fut.mean():,.1f} MW")
print("LSTM forecast done.")
```

## Conclusion and Recommendations

### Summary of Findings

This project successfully built, evaluated, and compared four time
series forecasting models on 5,057 daily electricity demand
observations from the AEP grid (2004-2018).

### Model Performance Summary

| Model | RMSE (MW) | MAE (MW) | MAPE (%) | R² |
|-------|-----------|----------|----------|----|
| ARIMA(3,1,3) | 2,453.98 | 1,885.90 | 11.89% | -0.8154 |
| SARIMA(2,1,0)(2,0,0,7) | 2,014.40 | 1,529.17 | 9.84% | -0.2233 |
| Prophet | — | — | 7.49% | — |
| **LSTM** | **736.10** | **571.54** | **3.86%** | **0.8367** |

### Key Findings

1. LSTM outperformed all other models on every metric, achieving
   96.14% accuracy (MAPE 3.86%) and explaining 83.67% of all
   demand variation (R² = 0.8367).

2. SARIMA improved ARIMA by capturing weekly seasonality (m=7),
   reducing MAPE from 11.89% to 9.84% — a 17.2% improvement.

3. Prophet performed well without any parameter tuning, achieving
   7.49% MAPE through automatic trend and seasonality detection.

4. The negative R² values for ARIMA and SARIMA indicate they
   performed worse than a naive mean forecast on this long horizon
   — highlighting the value of deep learning for complex patterns.

5. EDA revealed strong dual seasonality: summer peaks (air
   conditioning) and winter peaks (heating), with weekdays
   consistently 800+ MW higher than weekends.

### Practical Recommendations

For grid operators and energy planners:

1. Deploy LSTM for day-ahead and week-ahead demand forecasting
   given its superior accuracy on this dataset.

2. Maintain SARIMA as a fast, interpretable backup model for
   operational use when computational resources are limited.

3. Incorporate exogenous variables (temperature, humidity,
   economic indicators) as additional LSTM inputs to further
   reduce forecast error.

4. Use Prophet for rapid exploratory forecasting and
   communicating trends to non-technical stakeholders.

### Limitations

1. Only historical demand was used — weather data would
   significantly improve model accuracy.
2. The LSTM lookback window (60 days) was fixed — optimal
   window search could improve performance further.
3. Dataset ends in 2018 — post-COVID demand patterns may
   differ substantially.

### Future Work

- Incorporate temperature and weather features (multivariate LSTM)
- Test Transformer-based models (Temporal Fusion Transformer)
- Implement rolling walk-forward cross-validation
- Deploy as a real-time Streamlit forecasting dashboard
- Extend to regional sub-grid level forecasting
